```python
import json
import torch
import pandas as pd
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from tqdm.auto import tqdm
%env TOKENIZERS_PARALLELISM=true
# device = torch.device('cuda')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
class CFG:
    INPUT = '/kaggle/input/bengaliai-speech'
    TRANS_MODEL = 'facebook/nllb-200-distilled-600M'
```

```python
# df = pd.read_csv(f'{CFG.INPUT}/train.csv')
df = pd.read_csv('train.csv')
display(df)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(CFG.TRANS_MODEL)
torch.save(model.state_dict(), './nllb-200-distilled-600M.pth')
model.eval()
model.to(device)
tokenizer = AutoTokenizer.from_pretrained(CFG.TRANS_MODEL)
tokenizer.save_pretrained('./tokenizer/')
```

```python
def translate_by_nllb(text, tokenizer, model, device):
    inputs = tokenizer(text, return_tensors="pt")
    for k, v in inputs.items():
        inputs[k] = v.to(device)
    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"], max_length=64
    )
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    if i<5000:
        df.loc[i,'eng'] = translate_by_nllb(row['sentence'], tokenizer, model, device)
display(df[0:5000])
```

```python
import csv

# Read from input file and split by tabs
with open('ben.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Process each line
for line in lines:
    parts = line.strip().split('\t')
    if len(parts) >= 2:
        eng_text = parts[0].strip()
        ben_text = parts[1].strip()

        # Check if English text has 4 or more words
        if len(eng_text.split()) >= 4:
            # Write English text to eng_test.txt
            with open('eng_test.txt', 'a', encoding='utf-8') as eng_file:
                eng_file.write(eng_text + '\n')

            # Write Bengali text to ben_test.csv
            with open('ben_test.csv', 'a', newline='', encoding='utf-8') as ben_file:
                writer = csv.writer(ben_file, delimiter='\t')
                writer.writerow([ben_text])

print("Files eng_test.txt and ben_test.csv created successfully.")
```

```python
import pandas as pd
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from tqdm.auto import tqdm
import torch

# Load the pre-trained model and tokenizer
# model = AutoModelForSeq2SeqLM.from_pretrained('./nllb-200-distilled-600M.pth')
# model.eval()
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# model.to(device)
# tokenizer = AutoTokenizer.from_pretrained('./tokenizer/')

# Function to translate text
def translate_by_nllb(text, tokenizer, model, device):
    inputs = tokenizer(text, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    translated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"],
        max_length=64
    )
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    return translated_text

# Read Bengali text from ben_test.csv and translate
translated_texts = []
df_ben = pd.DataFrame(columns=['Bengali'])
with open('ben_test.csv', 'r', encoding='utf-8') as file:
    lines = file.readlines()
    for line in tqdm(lines, total=len(lines)):
        try:
            text = line.strip()
            translated_text = translate_by_nllb(text, tokenizer, model, device)
            translated_texts.append(translated_text)
        except Exception as e:
            print(f"Error processing line: {line}, Error: {e}")

# Save translated text to translate_test.txt
with open('translate_test.txt', 'w', encoding='utf-8') as file:
    for text in translated_texts:
        file.write(text + '\n')

print("Translation complete. Translated text saved to translate_test.txt")
```

#BERT Scores

```python
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# Load the pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Read sentences from files
with open('eng_test.txt', 'r') as file1, open('translate_test.txt', 'r') as file2:
    sentences1 = [line.strip() for line in file1.readlines()]
    sentences2 = [line.strip() for line in file2.readlines()]

# Initialize a list to store cosine similarities
bert_similarity_list = []

# Loop through pairs of sentences
for sent1, sent2 in zip(sentences1, sentences2):
    # Tokenize and prepare the texts for BERT
    inputs1 = tokenizer(sent1, return_tensors="pt", padding=True, truncation=True)
    inputs2 = tokenizer(sent2, return_tensors="pt", padding=True, truncation=True)

    # Feed the texts to the BERT model
    outputs1 = model(**inputs1)
    outputs2 = model(**inputs2)

    # Obtain the representation vectors
    embeddings1 = outputs1.last_hidden_state.mean(dim=1).detach().numpy()
    embeddings2 = outputs2.last_hidden_state.mean(dim=1).detach().numpy()

    # Calculate cosine similarity
    similarity = np.dot(embeddings1, embeddings2.T) / (np.linalg.norm(embeddings1) * np.linalg.norm(embeddings2))

    # Append the similarity score to the list
    bert_similarity_list.append(similarity[0][0])

    # Output the pairs with their scores
    # print("{} \t\t {} \t\t BERT Similarity: {:.4f}".format(sent1, sent2, similarity[0][0]))

# Print the list of BERT similarities
# print("BERT Similarities List:", bert_similarity_list)
print("===================================================================================")
avg_Bert_score = sum(bert_similarity_list)/len(bert_similarity_list)
print("Average Sentences BERT Scores: {:.4f}".format(avg_Bert_score))
print("===================================================================================")
```
